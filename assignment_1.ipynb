Task: load, display and do horizontal and vertical flip.

```python
import cv2
from matplotlib import pyplot as plt

# Read image and convert to RGB
img_bgr = cv2.imread("./images/lenna.png")
img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Flip image and concatenate all the images
img_all = cv2.hconcat([img, img[:, ::-1], img[::-1, :]])

# Save the image
cv2.imwrite("./images/assignment_1.png", cv2.cvtColor(img_all, cv2.COLOR_RGB2BGR))

plt.imshow(img_all)
plt.show()
```

The pixels of the array representing the image are inverted in X or Y order to return the image using python slicing:
array[start:stop:step]
